```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#to enable inline plotting
%matplotlib inline
```

```python
#step 1 Obtaining data
df = pd.read_csv('train.csv', encoding='latin-1')
```

```python
df.head()
```

```python
summary_stats = df[['SalePrice', 'TotRmsAbvGrd', 'OverallCond']].describe()
print(summary_stats)
```

```python
#Produce summary statistics for Traindata including salesprice
summary=df.describe()
print(summary)
```

```python
#Produce summary statistics for salesprice
summary_saleprice = df['SalePrice'].describe()
print(summary_saleprice)
```

```python
#Produce summary statistics for salesprice(TotalRms AbvGrd)
summary_TotRmsAbvGrd = df['TotRmsAbvGrd'].describe()
print(summary_TotRmsAbvGrd)
```

```python
#Produce summary statistics for salesprice(Overallcondition)
summary_OverallCond = df['OverallCond'].describe()
print(summary_OverallCond)
```

```python
# Visualizations for SalePrice, TotRmsAbvGrd, and OverallCond
plt.figure(figsize=(15, 5))

# Histogram for SalePrice
plt.subplot(1, 3, 1)
sns.histplot(df['SalePrice'], kde=True)
plt.title('SalePrice Distribution')

# Histogram for TotRmsAbvGrd
plt.subplot(1, 3, 2)
sns.histplot(df['TotRmsAbvGrd'], kde=True)
plt.title('TotRmsAbvGrd Distribution')

# Histogram for OverallCond
plt.subplot(1, 3, 3)
sns.histplot(df['OverallCond'], kde=True)
plt.title('OverallCond Distribution')

plt.tight_layout
plt.show()
```

```python
 #Interpretive Text
class color:
       BOLD = '\033[1m'
       END = '\033[0m'
print(color.BOLD + "Summary Statistics:" + color.END)
print(summary_stats)
print(color.BOLD + "\nInterpretation:"+ color.END)
print(color.BOLD + "Sale Price:"+ color.END)
print("The 'SalePrice' distribution is positively skewed(towards the right). This indicates that the more affordable houses are in high demand compared to the relatively costlier houses which are less in number")
print(color.BOLD + "\nTotal Rooms Above Grade:" + color.END)
print("The 'TotRmsAbvGrd' distribution is approximately normal, with the majority of houses having around 6-7 rooms above grade.")

print(color.BOLD + "\nOverall Condition:" + color.END)
print("The 'OverallCond' distribution indicates that the houses with the highest demand are those that have an overall condition rating close to 5, which is considered average. This can point at houses with a high overall condition tend to be expensive hence lower demand and sales while those with a low overall rating show have low demand and hence low sales")
print(color.BOLD + "NB" + color.END)
print("The normal distribution of the overall condition of the houses as well as Total rooms above grade is evidence that the data is reliable in terms of accuracy")
```

```python


#Explore Correlations

#convert non-numeric columns to numeric using ordinal encoding
df = df.apply(lambda col: col.astype('category').cat.codes if col.dtype == 'object' else col)
# Interpretive Text
class color:
       BOLD = '\033[1m'
       END = '\033[0m'
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Find the features with the highest positive and negative correlations with 'SalePrice'
strongest_positive_corr = correlation_matrix['SalePrice'].sort_values(ascending=False).head(5)
strongest_negative_corr = correlation_matrix['SalePrice'].sort_values(ascending=True).head(5)

# Print the positive correlated features
print(color.BOLD +"Features with the Strongest Positive Correlations:"+ color.END)
print(strongest_positive_corr)

# Print the negative correlated feature
print(color.BOLD + "Features with the Strongest Negative Correlations:"+ color.END)
print(strongest_negative_corr)


# Strongest Positive Correlations
plt.figure(figsize=(14, 10))
for i, feature in enumerate(strongest_positive_corr.index):
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(x=feature, y='SalePrice', data=df, color='purple')
    plt.title(f'{feature} vs. SalePrice')
   
   
    # Strongest Negative Correlations
plt.figure(figsize=(14, 10))
for i, feature in enumerate(strongest_negative_corr.index):
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(x=feature, y='SalePrice', data=df, color='purple')
    plt.title(f'{feature} vs. SalePrice')
```

```python
#new feature age
df['Age'] = df['YrSold'] - df['YearBuilt']
#new feature age
df['Age'] = df['YrSold'] - df['YearBuilt']
# Plot the relationship between Age and SalePrice
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Age', y='SalePrice', data=df)
plt.xlabel('Age')
plt.ylabel('Sale Price')
plt.title('Relationship between Age and Sale Price')
plt.show()
```

```python
#Relationship between Overall Condition and Sale Price:
plt.figure(figsize=(10, 6))
sns.boxplot(x='OverallCond', y='SalePrice', data=df, palette='viridis')
plt.xlabel('Overall Condition')
plt.ylabel('Sale Price')
plt.title('Relationship between Overall Condition and Sale Price')
plt.show()
```

```python
# KDE Plot for Age vs Sale Price
plt.figure(figsize=(10, 8))
sns.kdeplot(x=df['Age'], y=df['SalePrice'], cmap="Blues", shade=True)
plt.title('Age vs Sale Price Density')
plt.xlabel('Age of the House')
plt.ylabel('Sale Price')
plt.show()
```
